```python
def creat_df(file) :
    if os.path.exists(file):
        df_from_csv = pd.read_csv(file, encoding='utf-8', delimiter=';',low_memory=False)
        return df_from_csv
    else:
        print('FICHIER NON TROUVE', file)
        return pd.DataFrame()
```
