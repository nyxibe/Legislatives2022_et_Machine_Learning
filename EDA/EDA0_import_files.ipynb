```python
def remove_words_from_text(text, words_to_remove):
    for word in words_to_remove:
        text = text.replace(word, '')
    return text
```

```python
def trouver_fichiers_csv(repertoire):
    fichiers_csv = []

    for dossier_racine, sous_dossiers, fichiers in os.walk(repertoire):
        for fichier in fnmatch.filter(fichiers, '*.csv'):
            chemin_complet = os.path.join(dossier_racine, fichier)
            fichiers_csv.append(chemin_complet)

    return fichiers_csv
```

### ALIMENTER df_dataframe 

```python
# Spécifiez le chemin du répertoire que vous souhaitez parcourir
chemin_repertoire = '../../ProjetCommun/DATA_SOURCES/HCP'

# Appel de la fonction pour récupérer la liste des fichiers CSV
fichiers_csv_trouves = trouver_fichiers_csv(chemin_repertoire)

# Affichage des chemins des fichiers CSV trouvés
for file in fichiers_csv_trouves:
    com='' 
    depart = 0
    district = 0
    commune = 0
    canton = 0
    yyyy = 0
    if os.path.exists(file) :
        print (file)
        directory, filename = os.path.split(file)
        name, extension = os.path.splitext(filename)
        if extension == '.csv' : 
            if "commune" in name.lower():
                commune = 1
            if "departement" in name.lower():
                depart = 1
            if "district" in name.lower():
                depart = 1 
            if "canton" in name.lower():
                canton = 1 
            words_to_remove = ['departements','district','communes','cantons']    
            try:
                last4char = int(name[-4:])
                if last4char > 1750:
                    yyyy = last4char
                    words_to_remove.append(str(yyyy))
            except ValueError:
                yyyy =0
            cat = directory.split('/')[-2]
            com = remove_words_from_text(name, words_to_remove)
            new_data = {'category': cat, 
                       'filename': name, 
                       'comment' : com, 
                       'size': 0,
                       'depart': depart,
                       'district': district,
                       'commune': commune,
                       'canton': canton,
                       'yyyy': yyyy,
                       'path' : file}
            df_dataframe.loc[len(df_dataframe)] = new_data
    else:
        print(f"The CSV file '{file}' does not exist.")

```

```python
from pathlib import Path

def df_creat(filename, df_dataframe):  
    df = df_dataframe[df_dataframe['filename'] == filename]
    if len(df) == 1:
        row = df.iloc[0]
        file = row['path']
        if Path(file).exists() :
            df_from_csv = pd.read_csv(file, encoding='utf-8', low_memory=False)
            print ('DATAFRAME SIZE : ',df_from_csv.shape)
            return df_from_csv
        else:
            print('FICHIER NON TROUVE', filename)
            return pd.DataFrame()
    else:
        print(f"filename: '{filename}' return size: {len(df)}")
        return pd.DataFrame()

```

```python
def col_XLSX (i,i0=65) :
    l = chr(i % i0 + i0)
    if i > 90 :
        l = 'A' + chr((i-26) % i0 + i0)
    return l
col_XLSX (113)
```
