```python
import warnings
warnings.filterwarnings("ignore")

import os
import zipfile
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def creat_df(file) :
    if os.path.exists(file):
        df_from_csv = pd.read_csv(file, encoding='utf-8', delimiter=';',low_memory=False)
        return df_from_csv
    else:
        print('FICHIER NON TROUVE', file)
        return pd.DataFrame()
```

```python
file_path2="../../ProjetCommun/DATA_SOURCES/Others/correspondance-code-insee-code-postal.csv"
```

```python
df2 = creat_df(file_path2) 
df2.head(10)
```

```python
df2[df2['Code INSEE']=='78575']
```

```python
df2['dep'] = df2['Code Département']
df2['nomdep'] = df2['Département'].str.strip('[\'\']')  
df2[df2['Code INSEE']=='78575']
```

```python
df2['codecommune'] = df2['Code INSEE'].astype(str).str.zfill(5)
df2['nomcommune'] = df2['Commune'] 
df2[df2['Code INSEE']=='78575']
```

```python
df2[['GEO_latitude', 'GEO_longitude']] = df2['geo_point_2d'].str.split(', ', expand=True)
df2[df2['Code INSEE']=='78575']
```

```python
df2[['GEO_latitude', 'GEO_longitude']] = df2[['GEO_latitude', 'GEO_longitude']].apply(pd.to_numeric)
df2['DBL_geo_point_2d'] = df2['geo_point_2d'] 
df2['DBL_coderegion'] = df2['Code Région'] 
df2['DBL_nomregion'] = df2['Région'].str.strip('[\'\']') 
df2['DBL_codepostale'] = df2['Code Postal'] 
df2[df2['Code INSEE']=='78575']
```

## SQL 

```python
df4_sql=df2[
    ['codecommune','DBL_codepostale',
     'nomcommune','nomdep','DBL_nomregion',
     'GEO_latitude','GEO_longitude',
     'Altitude Moyenne','Superficie',
     'Code Département','DBL_coderegion']]
df4_sql['codecommune'] = df4_sql['codecommune'].astype(str).str.zfill(5)
df4_sql['key']=df4_sql['codecommune']
df4_sql[df4_sql['codecommune']=='78575']
```

```python
df4_sql.set_index('key', inplace=True)
df4_sql[df4_sql['codecommune']=='78575']
```

```python
for i,c in enumerate(df4_sql) :
    print (i,c)
```

```python
new_column_names = {
        'DBL_codepostale': 'codepostale',
        'DBL_nomregion': 'nomregion',
        'GEO_latitude': 'latitude',
        'GEO_longitude': 'longitude',
        'Altitude Moyenne': 'altitude',
        'Superficie': 'superficie',
        'Code Département': 'codedepartement',        
        'DBL_coderegion': 'coderegion'}
df4_sql.rename(columns=new_column_names, inplace=True)
df4_sql[df4_sql['codecommune']=='78575']
```

```python
df4_sql = df4_sql.drop(df4_sql.columns[0], axis=1)
```

```python
df4_sql[df4_sql['codepostale']=='89770']
```

```python
from sqlalchemy import create_engine, Column, Integer, String, MetaData, Table, update, select
db_path = '../DATA/HistoPlotik.db'
engine = create_engine(f'sqlite:///{db_path}')
metadata = MetaData()

tbl_db_names = metadata.tables.keys()
tbl_db_names
```

```python
# CREATION 
tbl_data = 'villes_tbl'
if tbl_data not in tbl_db_names :
    # Convert DataFrame to a SQL table, using the index as the ID column
    df4_sql.to_sql(tbl_data , con=engine, index=True, if_exists='replace')
    metadata.reflect(bind=engine)
tbl_db_names = metadata.tables.keys()
tbl_db_names
```
