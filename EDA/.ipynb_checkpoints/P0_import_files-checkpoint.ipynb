```python
import fnmatch

def trouver_fichiers_csv(repertoire):
    fichiers_csv = []

    for dossier_racine, sous_dossiers, fichiers in os.walk(repertoire):
        for fichier in fnmatch.filter(fichiers, '*.csv'):
            chemin_complet = os.path.join(dossier_racine, fichier)
            fichiers_csv.append(chemin_complet)

    return fichiers_csv

# Spécifiez le chemin du répertoire que vous souhaitez parcourir
chemin_repertoire = "../DATA_SOURCES/HCP"

# Appel de la fonction pour récupérer la liste des fichiers CSV
fichiers_csv_trouves = trouver_fichiers_csv(chemin_repertoire)

# Affichage des chemins des fichiers CSV trouvés
lg_char_analys = len('.csv')
suppr='.csv'
for file in fichiers_csv_trouves:
    com='' 
    depart = 0
    district = 0
    commune = 0
    canton = 0
    yyyy = 0
    file_size = os.path.getsize(file) / (8 * 10**6)
    directory, filename = os.path.split(file)
    if lg_char_analys == 0 or file[:lg_char_analys] != suppr:
        directory, filename = os.path.split(file)
        name, extension = os.path.splitext(filename)
        if extension == '.csv' : 
            print (file)
            if "comm" in name.lower():
                commune = 1  
            try:
                last4char = int(name[-8:-4])
                if last4char > 1750:
                    yyyy = last4char
            except ValueError:
                yyyy = 0
            com = remove_words_from_text(name, words_to_remove)
            new_data = {'category': directory.split('/')[-2], 
                           'filename': name, 
                           'comment' : com, 
                           'size': file_size,
                           'depart': depart,
                           'district': district,
                           'commune': commune,
                           'canton': canton,
                           'yyyy': yyyy,
                           'path' : file}
            df.loc[len(df)] = new_data



```

```python
def df_creat(filename, df_dataframe):  
    df = df_dataframe[df_dataframe['filename'] == filename]
    if len(df) == 1:
        row = df.iloc[0]
        if row['category'] == 'leg':
            zip_file_path = "../DATA/Legislatives/leg" + str(row['yyyy']) + '_csv.zip'
            csv_file_path = 'leg' + str(row['yyyy'])+ '_csv/leg' + str(row['yyyy']) + 'comm.csv'
            removdir = "../DATA/Legislatives/leg" + str(row['yyyy'])+ '_csv' 
            print(zip_file_path , csv_file_path,removdir)
        else:
            zip_file_path = "../DATA/" + row['category'] + '.zip'
            csv_file_path = row['category'] + '/' + filename + '.csv'
            removdir = "../DATA/" + row['category']
        if os.path.exists(zip_file_path):
            with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
                file_list = zip_ref.namelist()
                if csv_file_path in file_list:
                    # UNZIP FILE to csv
                    zip_ref.extract(csv_file_path, path=zip_file_path[:-4])                  
                    # CREATE DATAFRAME WITH CSV
                    out=zip_file_path[:-4] +'/'+ csv_file_path
                    print ('OUT >>',out)
                    df_from_csv = pd.read_csv(out, encoding='utf-8')
                    print ('DATAFRAME SIZE : ',df_from_csv.shape)
                    # DELETE CSV FILE and PARENT DIRECTORIES
                    # os.remove(out)
                    # os.rmdir(os.path.dirname(out))
                    # os.rmdir(removdir)
                    return df_from_csv
                else: 
                    print(csv_file_path, 'NOT in', file_list)
        else:
            print('FICHIER NON TROUVE', zip_file_path)
            return pd.DataFrame()
    else:
        print(f"filename: '{filename}' return size: {len(df)}")
        return pd.DataFrame()

```

```python
def col_XLSX (i,i0=65) :
    l = chr(i % i0 + i0)
    if i > 90 :
        l = 'A' + chr((i-26) % i0 + i0)
    return l
col_XLSX (113)
```

```python
import random
import openpyxl
# Create a new workbook
workbook = openpyxl.Workbook()
# Create sheets
sheet1 = workbook.active  # Default sheet created with the workbook
sheet1.title = "BIBLIO"  # Rename the default sheet
sheet2 = workbook.create_sheet(title="INFOS")
sheet3 = workbook.create_sheet(title="Exemple_x2")
xc = 65 # 'A'
xl = 1
list_codecommune=[]
for index, row in filtered_df.iterrows():
    # Construct the DataFrame name dynamically
    df_name = 'df_' + row['filename']
    try:
        df = df_creat(row['filename'], df_dataframe)
        print (row['filename'],df.shape)
        if df.shape[0] > 0 :
            # sheet1
            sheet1['A'+str(xl)] = row['filename']
            sheet1['B'+str(xl)] = row['category']  
            sheet1['C'+str(xl)] = df.shape[0]  
            sheet1['D'+str(xl)] = df.shape[1]  
            sheet1['E'+str(xl)] = row['comment']
            sheet1['F'+str(xl)] = str(row['yyyy'])
            print (xl, row['filename'], 'ok')
            xl+=1
            # sheet2 & sheet3
            xcol = col_XLSX (xc)
            sheet2[xcol+'1'] = row['filename']
            sheet3[xcol+'1'] = row['filename']
            sheet2[xcol+'2'] = row['category']
            sheet3[xcol+'2'] = row['category']
            l=5
            for c in df :
                tot = df[c].nunique()
                totNULL = df[c].isnull().sum()
                NULLp100 = round(totNULL/df.shape[0]*100,2)
                # val=df.iloc[c, random.randint(0, df.shape[1] - 1)]
                val1 = df[c].iloc[random.randint(0, df.shape[0] - 1)]
                val2 = df[c].iloc[random.randint(0, df.shape[0] - 1)]
                if c == 'codecommune' :
                    list_codecommune.append(row['filename'])
                try:
                    pd.to_numeric(df[c])
                    mini = df[c].min()
                    maxi = df[c].max()
                except ValueError:
                    mini = ''
                    maxi = ''
                sheet2[xcol+str(l)] = c+' :'+str(tot)+' ('+str(totNULL)+' '+str(NULLp100)+'%) ['+str(mini)+';'+str(maxi)+']'
                try:
                    sheet3[xcol+str(l)] = c+' : '+str(val1)+' - '+str(val2)
                except ValueError:
                    sheet3[xcol+str(l)] = 'ENCODING ERROR'
                l+=1    
            xc+=1
            if xl>90:
                break 
        else : 
            print (row['filename'] ,'0 ligne') 
            sheet1['A'+str(xl)] = row['filename']
            sheet1['B'+str(xl)] = '0 ligne'    
            xl+=1  
    except ValueError:
        print ('CREATION DATAFRAME IMPOSSIBLE FOR',row['filename'])
        sheet1['A'+str(xl)] = row['filename']
        sheet1['B'+str(xl)] = 'CREATION DATAFRAME IMPOSSIBLE !!!'    
        xl+=1   
# Save the workbook
workbook.save('Catalogue_DF_select.xlsx')
```
