## <a name="intro"></a>Intro : Import & functions&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<a href="#TOP">&uarr;</a>

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import os
import zipfile
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def df_creat(filename, df_dataframe):  
    df = df_dataframe[df_dataframe['filename'] == filename]
    if len(df) == 1:
        row = df.iloc[0]
        if row['category'] == 'leg':
            zip_file_path = "../DATA/Legislatives/leg" + str(row['yyyy']) + '_csv.zip'
            csv_file_path = 'leg' + str(row['yyyy'])+ '_csv/leg' + str(row['yyyy']) + 'comm.csv'
            removdir = "../DATA/Legislatives/leg" + str(row['yyyy'])+ '_csv' 
            print(zip_file_path , csv_file_path,removdir)
        else:
            zip_file_path = "../DATA/" + row['category'] + '.zip'
            csv_file_path = row['category'] + '/' + filename + '.csv'
            removdir = "../DATA/" + row['category']
        if os.path.exists(zip_file_path):
            with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
                file_list = zip_ref.namelist()
                if csv_file_path in file_list:
                    # UNZIP FILE to csv
                    zip_ref.extract(csv_file_path, path=zip_file_path[:-4])                  
                    # CREATE DATAFRAME WITH CSV
                    out=zip_file_path[:-4] +'/'+ csv_file_path
                    print ('OUT >>',out)
                    df_from_csv = pd.read_csv(out, encoding='utf-8')
                    # print ('DATAFRAME SIZE : ',df_from_csv.shape)
                    # DELETE CSV FILE and PARENT DIRECTORIES
                    # os.remove(out)
                    # os.rmdir(os.path.dirname(out))
                    # os.rmdir(removdir)
                    return df_from_csv
                else: 
                    print(csv_file_path, 'NOT in', file_list)
        else:
            print('FICHIER NON TROUVE', zip_file_path)
            return pd.DataFrame()
    else:
        print(f"filename: '{filename}' return size: {len(df)}")
        return pd.DataFrame()
```

```python
csv_file_path = 'df_filtred.csv'
# Use pandas to read the CSV file into a DataFrame with UTF-8 encoding
df_cat = pd.read_csv(csv_file_path, encoding='utf-8')
```

## <a name="ImportZip"></a>Partie 2 : CREATION df_exemples &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<a href="#TOP">&uarr;</a>

```python
df_filtred = df_cat[df_cat['codecommune']==1]
df_filtred.shape
```

```python
dataframes = {}
for file_name in df_cat['filename'].tolist() :
    print (file_name)
    df_name = 'df_' + file_name
    try:
        df = df_creat(file_name, df_filtred)
        size_full = df.shape
        df_cat.loc[df_cat['filename'] == file_name, 'col'] = df.shape[1]
        df_cat.loc[df_cat['filename'] == file_name, 'l_full'] = df.shape[0]
        if 'codecommune' in df.columns:
            length_difference = len(set(target_codecommune) - set(df['codecommune'].unique()))
            df_cat.loc[df_cat['filename'] == file_name, 'nb_commun_absent'] = length_difference
            print ('MANQUE ', length_difference)
            df = df[(df['codecommune'] == 78575) | (df['codecommune'] == 89069)]
            size_2citys = df.shape
            df_cat.loc[df_cat['filename'] == file_name, 'l_2exempl'] = df.shape[0]
        else :
            size_2citys = [0,0]
        dataframes[file_name] = df
    except ValueError:
        print ('ERREUR A LA CREATION')
```
