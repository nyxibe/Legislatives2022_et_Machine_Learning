## <a name="Import"></a>Partie 1 : Import CSV&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<a href="#TOP">&uarr;</a>

```python
df = pd.read_csv('diplomescommunes.csv', encoding='utf-8')
df.shape
```

```python
df.columns
```

```python
colonnes_non_nulles = df.columns[df.notna().all()]

for colonne in colonnes_non_nulles:
    print(colonne)
```

```python
df.head()
```

## <a name="Import"></a>Partie 2 : Renommer &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<a href="#TOP">&uarr;</a>

```python
df['DBL_dep'] = df.dep
df['DBL_nomdep'] = df.nomdep
df['DBL_nomcommune'] = df.nomcommune
```

```python
df['codecommune'] = df['codecommune'].astype(str).str.zfill(5)
```

```python
file_path = '../DATA/list_codecommune_leg2022_50inscrits.txt'
with open(file_path, 'r', encoding='utf-8') as file:
    lines = [line.strip() for line in file.readlines()]

print(lines[0:10], len(lines))
```

```python
codecommune_listdf = df['codecommune'].tolist()
```

```python
absent = [x for x in lines if x not in codecommune_listdf]
len (absent)
```

```python
df_filtered = df[df['codecommune'].isin(lines)]
```

```python
df_filtered.shape
```

```python
col = []
for c in df_filtered.columns:
  if 'bach' in c or 'bacf' in c or 'suph' in c or 'supf' in c:
    col.append(c)

colbis = []
for c in col :
  if '2022' in c or '2012' in c or '2002' in c or '1992' in c or '2017' in c:
    colbis.append(c)

colbis
```

info : les codes communes de df_50 commencent par un 0 et non un 1

```python
df_final = df_filtered[colbis]
```

```python
df_final.columns = 'DIP_' + df_final.columns
```

```python
df_final.shape
```

```python
df_final = pd.concat([df_filtered[['dep', 'nomdep', 'codecommune', 'nomcommune']], df_final], axis = 1)
```

```python
df_final['DBL_bug'] = 0
df_final.shape
```

```python
df_final.head(2)
```

```python
new_row = ['*', '*', 'code', '*'] + [np.nan] * 20+[1]
len (new_row)
```

```python
for i,code in enumerate(absent) :
    new_row = ['*', '*', code, '*'] + [np.nan] * 20 + [1]
    row_df = pd.DataFrame([new_row], columns=df_final.columns)
    df_final = pd.concat([df_final, row_df], ignore_index=True)
df_final.shape
```

## <a name="Import"></a>Partie 3 : Export CSV&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;<a href="#TOP">&uarr;</a>

```python
df_final.to_csv('../DATA/EXPORT_CSV/final_DIP_diplomescommunes.csv', sep=';', index=False)
```
